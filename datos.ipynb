```python
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np


estadisticas = pd.read_csv("estadisticas.csv", sep = ",")
estadisticas
estadisticas.columns = ['Tiempo','', 'CPU', 'Memoria Usada', 'Memoria Libre']


        
        

estadisticas['Tiempo'] = pd.to_datetime(estadisticas['Tiempo'])
                                                    

# Crear los gráficos
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 8), sharex=True)

# Grafico de uso de CPU
axes[0].plot(estadisticas['Tiempo'], estadisticas['CPU'], label='Uso de CPU', color='blue')
axes[0].set_ylabel('Uso de CPU (%)')
axes[0].set_title('Uso de CPU a lo largo del tiempo')
axes[0].legend()

# Grafico de memoria usada
axes[1].plot(estadisticas['Tiempo'], estadisticas['Memoria Usada'], label='Memoria Usada', color='green')
axes[1].set_ylabel('Memoria Usada (MB)')
axes[1].set_title('Memoria Usada a lo largo del tiempo')
axes[1].legend()

# Grafico de memoria libre
axes[2].plot(estadisticas['Tiempo'], estadisticas['Memoria Libre'], label='Memoria Libre', color='red')
axes[2].set_xlabel('Tiempo')
axes[2].set_ylabel('Memoria Libre (MB)')
axes[2].set_title('Memoria Libre a lo largo del tiempo')
axes[2].legend()

# Formato del eje x
for ax in axes:
    ax.xaxis.set_major_locator(mdates.SecondLocator(interval=35))  # Marcas cada 5 segundos
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    ax.tick_params(axis='x', rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad

# Ajustar el layout para que los subplots no se solapen
fig.tight_layout()

# Mostrar el gráfico
plt.show()
```
